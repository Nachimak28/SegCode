```python
#1
#local and global thresholding

from scipy.misc import imread, imsave, imresize
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np
```

```python
img = cv2.imread('trial_2.png', 0)
```

```python
img.shape
```

```python
#global thresholding
med = []
for i in range(len(img)):
    for j in range(0,len(img[i]), 25):
        med.append(img[i][j])
        

#len(med)

S = np.median(med)

print(S)

#med

dev = []
for i in range(len(img)):
    for j in range(len(img[i])):
        dev.append(abs(img[i][j] - S))

#len(dev)

#dev

sd = np.median(dev)

print(sd)

alpha = 1    #change to higher values for different results
T = S - alpha * sd

print(T)

plt.plot(np.arange(img[150].size), img[150], "ro")

cp = img.copy()
for i in range(len(cp)):
    for j in range(len(cp[i])):
        if cp[i][j] <= T:
            cp[i][j] = 255

plt.title('Global Thresholding')
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap  = 'gray')
plt.subplot(1,2,2)
plt.title('Global Thresholding')
plt.imshow(cp, cmap = 'gray')
```

<h1>Hybrid gray scale models</h1>

```python
kernel = np.array([[0,0,-1,0,0],[0,0,-2,0,0],[-1,-2,0,2,1],[0,0,2,0,0],[0,0,1,0,0]])

import scipy.signal

edges = scipy.signal.convolve2d(img, kernel, 'valid')

plt.imshow(edges, cmap = 'gray')
```

```python
plt.plot(np.arange(295), edges, "o")
```

```python
uR = np.mean(edges)
sdR = np.std(edges)
alpha = 1.5
T = uR + alpha * sdR
print(T)
```

```python
cp = edges.copy()

for i in range(len(cp)):
    for j in range(len(cp[i])):
        if cp[i][j] >= T:
            cp[i][j] = 255
            
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap  = 'gray')
plt.subplot(1,2,2)
plt.title('Hybrid Gray scale model with structural features(Global thresholding)')
plt.imshow(cp, cmap = 'gray')
```

```python
cp1 = cv2.imread(edges,0)
th2 = cv2.adaptiveThreshold(cp1,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(cp1,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
```

```python
kernel = np.array([[-1,-2,0,2,1],[-1,-2,0,2,1],[-1,-2,0,2,1],[-1,-2,0,2,1],[-1,-2,0,2,1]])

import scipy.signal

edges = scipy.signal.convolve2d(img, kernel, 'valid')

plt.imshow(edges, cmap = 'gray')
```

```python
cp = edges.copy()

for i in range(len(cp)):
    for j in range(len(cp[i])):
        if cp[i][j] >= T:
            cp[i][j] = 255
            
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap  = 'gray')
plt.subplot(1,2,2)
plt.title('Hybrid Gray scale model with structural features(Global thresholding)')
plt.imshow(cp, cmap = 'gray')
```
