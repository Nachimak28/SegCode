```python
#1
#local and global thresholding

from scipy.misc import imread, imsave, imresize
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np
```

```python
img = cv2.imread('trial_2.png', 0)
```

```python
img.shape
```

```python
#global thresholding
med = []
for i in range(len(img)):
    for j in range(0,len(img[i]), 25):
        med.append(img[i][j])
        

#len(med)

S = np.median(med)

print(S)

#med

dev = []
for i in range(len(img)):
    for j in range(len(img[i])):
        dev.append(abs(img[i][j] - S))

#len(dev)

#dev

sd = np.median(dev)

print(sd)

alpha = 1    #change to higher values for different results
T = S - alpha * sd

print(T)

plt.plot(np.arange(img[150].size), img[150], "ro")

cp = img.copy()
for i in range(len(cp)):
    for j in range(len(cp[i])):
        if cp[i][j] <= T:
            cp[i][j] = 255

plt.title('Global Thresholding')
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap  = 'gray')
plt.subplot(1,2,2)
plt.title('Global Thresholding')
plt.imshow(cp, cmap = 'gray')
```

<h1>Hybrid gray scale models</h1>

```python
kernel = np.array([[0,0,-1,0,0],[0,0,-2,0,0],[-1,-2,0,2,1],[0,0,2,0,0],[0,0,1,0,0]])

import scipy.signal

edges = scipy.signal.convolve2d(img, kernel, 'valid')

plt.imshow(edges, cmap = 'gray')
```

```python
plt.plot(np.arange(295), edges, "o")
```

```python
uR = np.mean(edges)
sdR = np.std(edges)
alpha = 1.5
T = uR + alpha * sdR
print(T)
```

```python
cp = edges.copy()

for i in range(len(cp)):
    for j in range(len(cp[i])):
        if cp[i][j] >= T:
            cp[i][j] = 255
            
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap  = 'gray')
plt.subplot(1,2,2)
plt.title('Hybrid Gray scale model with structural features(Global thresholding)')
plt.imshow(cp, cmap = 'gray')
```

```python
import numpy as np
import cv2
 
def build_filters():
    filters = []
    ksize = 31
    for theta in np.arange(0, np.pi, np.pi / 16):
        kern = cv2.getGaborKernel((ksize, ksize), 4.0, theta, 10.0, 0.5, 0, ktype=cv2.CV_32F)
        kern /= 1.5*kern.sum()
        filters.append(kern)
    return filters
 
def process(img, filters):
    accum = np.zeros_like(img)
    for kern in filters:
        fimg = cv2.filter2D(img, cv2.CV_8UC3, kern)
    np.maximum(accum, fimg, accum)
    return accum
 
if __name__ == '__main__':
    import sys
 
    print __doc__
    try:
        img_fn = sys.argv[1]
    except:
        img_fn = 'test.png'
 
    img = cv2.imread(img_fn)
    if img is None:
    print 'Failed to load image file:', img_fn
 sys.exit(1)
 
 filters = build_filters()
 
 res1 = process(img, filters)
 cv2.imshow('result', res1)
 cv2.waitKey(0)
 cv2.destroyAllWindows()
```
