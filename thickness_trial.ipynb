```python
n = len(b)
```

```python
mean = sum(b * y_pos)/ n
```

```python
sigma = sum(y_pos*(b-mean)**2)/n
```

```python
from scipy.optimize import curve_fit
```

```python
def gaus(x,a,x0,sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))
```

```python
popt,pcov = curve_fit(gaus,b,y_pos,p0=[1,mean,sigma])
```

```python
plt.plot(y_pos,b,'b+:',label='data')
plt.plot(gaus(b,*popt),b,'ro:',label='fit')
plt.legend()
plt.show()
```

```python
#final algo using median
import time
tic = time.time()
img = cv2.imread('C:/Users/nachiket/Desktop/RAYD8 tech/Reference material/Segmentation codebase/trial_2.png', 0)
m = np.zeros_like(img)
v_img = cv2.flip(img, 1)
n = np.zeros_like(v_img)

c = 15

for i in range(len(img)):
    for j in range(len(img[i])):
         if img[i][j] <= np.median(img[i][j:j+c]):
                m[i][j] = 100

for i in range(len(v_img)):
    for j in range(len(v_img[i])):
        if v_img[i][j] <= np.median(v_img[i][j:j+c]):
            n[i][j] = 100
            
f_img = cv2.flip(n, 1)
res = np.concatenate((f_img[:,:100], m[:,65:]), axis = 1)
toc = time.time()
print('Time taken: ', str((toc - tic)), 's')
plt.imshow(res, cmap = plt.cm.gray)
plt.axis('off')
plt.show()
```
